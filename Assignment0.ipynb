```
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
# Reading an image in OpenCV
#drive.mount('/content/drive')
img1 = cv2.imread('/usr/image/img1.jpeg')
img2 = cv2.imread('/usr/image/img2.jpeg')
img3 = cv2.imread('/usr/image/img3.jpeg')
img4 = cv2.imread('/usr/image/img4.jpeg')

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
img4 = cv2.cvtColor(img4, cv2.COLOR_BGR2RGB)
plt.figure(figsize = [10,10])
#plt.axis('off') # Remove the axes of the plot

plt.subplot(2,2,1)
plt.imshow(img1)
plt.subplot(2,2,2)
plt.imshow(img2)
plt.subplot(2,2,3)
plt.imshow(img3)
plt.subplot(2,2,4)
plt.imshow(img4)
print(img1)
```
